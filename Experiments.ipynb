```python
#shows the response of the neural system on different stimulus
def dynamicNeuralSystemPlot():
    e = 0
    data = plotPattern(e, 1, 221, 'b')
    xk = data[50:250]
    xkmo = data[49:249]
    plt.figure(2)
    plt.plot(xk, xkmo, 'b')

    e = 0.7
    data = plotPattern(e, 1, 222, 'g')
    xk = data[50:250]
    xkmo = data[49:249]
    plt.figure(2)
    plt.plot(xk, xkmo, 'g')

    e = 0.9
    data = plotPattern(e, 1, 223, 'r')
    xk = data[50:250]
    xkmo = data[49:249]
    plt.figure(2)
    plt.plot(xk, xkmo, 'r')

    e = 1
    data = plotPattern(e, 1, 224, 'k')
    xk = data[50:250]
    xkmo = data[49:249]
    plt.figure(2)
    plt.plot(xk, xkmo, 'k')
    
    plt.title('Neural system response on diferent stimulus')
    plt.xlabel('X(k)')
    plt.ylabel('X(k-1)')
```

```python
dynamicNeuralSystemPlot()
```
